Quartile Information

```python
print('Shape of the data:',df.shape)

df=df[df['price']<=14000000]
# There seems to be two data points which hint towards being outliers

mean_power = df[df['type']=='power']['price'].mean()
mean_sail = df[df['type']=='sail']['price'].mean()
mean_unpowered = df[df['type']=='unpowered']['price'].mean()

median_power = df[df['type']=='power']['price'].median()
median_sail = df[df['type']=='sail']['price'].median()
median_unpowered = df[df['type']=='unpowered']['price'].median()


Q1_power = df[df['type']=='power']['price'].quantile(q=0.25)
Q1_sail = df[df['type']=='sail']['price'].quantile(q=0.25)
Q1_unpowered = df[df['type']=='unpowered']['price'].quantile(q=0.25)

Q3_power = df[df['type']=='power']['price'].quantile(q=0.75)
Q3_sail = df[df['type']=='sail']['price'].quantile(q=0.75)
Q3_unpowered = df[df['type']=='unpowered']['price'].quantile(q=0.75)


IQR_power = Q3_power - Q1_power
IQR_sail = Q3_sail - Q1_unpowered
IQR_unpowered = Q3_unpowered - Q1_unpowered


print(f'For the Powered\nMean: {mean_power} \nMedian: {median_power} \nFirst Quartile: {Q1_power} \nThird Quartile: {Q3_power} \nIQR: {IQR_power}\n\n')
print(f'For the Sail\nMean: {mean_sail} \nMedian: {median_sail} \nFirst Quartile: {Q1_sail} \nThird Quartile: {Q3_sail} \nIQR: {IQR_sail}\n\n')
print(f'For the Unpowered\nMean: {mean_unpowered} \nMedian: {median_unpowered} \nFirst Quartile: {Q1_unpowered} \nThird Quartile: {Q3_unpowered} \nIQR: {IQR_unpowered}')

lower_fence_power = Q1_power - 1.5*IQR_power
lower_fence_sail = Q1_sail - 1.5*IQR_sail
lower_fence_unpowered = Q1_unpowered - 1.5*IQR_unpowered

upper_fence_power = Q3_power + 1.5*IQR_power
upper_fence_sail = Q3_sail + 1.5*IQR_sail
upper_fence_unpowered = Q3_unpowered + 1.5*IQR_unpowered



out = df[(df[df['type']=='power']['price'] < lower_fence_power) | (df[df['type']=='power']['price'] > upper_fence_power)|(df[df['type']=='sail']['price'] < lower_fence_sail) | (df[df['type']=='sail']['price']> upper_fence_sail)|(df[df['type']=='unpowered']['price'] < lower_fence_unpowered) | (df[df['type']=='unpowered']['price']> upper_fence_unpowered)]

print(out)
```
