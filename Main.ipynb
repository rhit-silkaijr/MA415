Bringing in the Data

```python
df = pd.read_csv('ExoplanetArchiveAllClean.csv')
df.head(1)
df['pl_rade'] = pd.to_numeric(df['pl_rade'], errors='coerce')

df['logRadius'] = np.log(df['pl_rade'])
```

```python
a = df.shape
print('The Shape of the data is ', a)
```

```python
print('The following are how many particular instances of how many different exoplanets are contained in the inital dataset:')
df.value_counts(df['pl_name'])
```

```python
results_list = []
for features in df.columns:
   
    ratio = (df[features].isna() | (df[features] == "")).sum() / len(df)
    results_list.append({'f': features, 'ratio_empty': ratio})

# Create DataFrame once at the end
results = pd.DataFrame(results_list)

results = results.sort_values(by='ratio_empty', ascending=True)
results.head()

```

Histogram of the Distribution of the Regressor Target


```python
mean_val = df['logRadius'].mean()
median_val = df['logRadius'].median()

# 1. Create histogram
fig2 = px.histogram(df, x='logRadius', nbins=15, title='LogRadius Distribution')

# 2. Use add_vline instead of add_line, and add annotations
fig2.add_vline(x=mean_val, line_width=2, line_dash='dash', line_color='red',
               annotation_text=f'Mean: {mean_val:.1f}', annotation_position='top right')

fig2.add_vline(x=median_val, line_width=2, line_dash='solid', line_color='green',
               annotation_text=f'Median: {median_val:.1f}', annotation_position='top left')

# 3. Correct legend/layout positioning
fig2.update_layout(showlegend=True)

fig2.show()
```
